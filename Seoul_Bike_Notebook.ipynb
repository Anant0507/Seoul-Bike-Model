Categorical Analysis of univariate

1) Seasons

Count of all seasons stored in dictionary 

```python
Seasons_count = {}
Seasons_list = list(df["Seasons"].unique())
for i in Seasons_list:
    Seasons_count[i] = len(df[df["Seasons"] == i])
print(Seasons_count)
```

```python
sizes = list(Seasons_count.values())
labels = list(Seasons_count.keys())

plt.title("Seasons")
plt.pie(sizes, labels= labels, autopct="%1.2f%%")
```

Rented Bikes Count in Holidays

```python
Rented_Bike_Holiday_Total = {}
Holiday_Type = df["Holiday"].unique()
for i in Holiday_Type:
    Rented_Bike_Holiday_Total[i] = df[df["Holiday"] == i]["Rented Bike Count"].sum()
Rented_Bike_Holiday_Total
```

Holidays in different Seasons

```python
Holiday_Seasons_Total = {}
Season_Type = df["Seasons"].unique()
Holiday_Type = df["Holiday"].unique()


for i in Season_Type:
    for j in Holiday_Type:
        # print(df[(df["Seasons"] == i) and (df["Holiday"] == j)])
        Holiday_Seasons_Total[f'{i}_{j}'] = len(df[(df["Seasons"] == i) & (df["Holiday"] == j)])

Holiday_Seasons_Total
```

```python
x_values = list(Holiday_Seasons_Total.keys())
y_values = list(Holiday_Seasons_Total.values())
colors = ['red', 'blue', 'green', 'purple', 'orange', 'lime', 'cyan', 'magenta']

plt.bar(x_values, y_values, color = colors)

plt.xlabel('Seasons and Holidays')
plt.ylabel('Count of Holidays and No Holidays')
plt.title('Holidays in each Seasons')

plt.xticks(rotation=45, ha="right")

plt.show()
```

```python
sizes = list(Holiday_Seasons_Total.values())
labels = list(Holiday_Seasons_Total.keys())

plt.pie(sizes, labels = labels, autopct="%1.2f%%")
plt.show()
```
